Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

```python

import matplotlib.pyplot as plt

# Creating a simple time series dataset
np.random.seed(42)
dates = pd.date_range('2023-01-01', periods=100)
stock_prices = pd.Series(np.random.randint(50, 200, size=100), index=dates)

# Calculating the rolling mean using a window size of 10 days
rolling_mean = stock_prices.rolling(window=10).mean()


```

```python
# Plotting the original data and rolling mean
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, label='Stock Prices')
plt.plot(rolling_mean, label='Rolling Mean (Window=10)')
plt.title('Stock Prices and Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

In Pandas, window functions are a powerful tool for performing various operations on rolling or expanding windows of data in a DataFrame or Series. These functions allow you to apply calculations, aggregations, and transformations over a specified window of data, which can slide or expand along the data axis.

Window functions typically involve:

Partitioning: Dividing data into groups based on specified criteria.
Ordering: Arranging data within each partition in a defined order.
Window Specification: Defining the window size and other parameters.

rolling(): Creates a rolling object for performing calculations.
mean(), sum(), min(), max(): Compute mean, sum, minimum, maximum, etc. within the rolling window.
std(), var(): Calculate the standard deviation and variance within the window.

Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

```python
def date_difference(start_date, end_date):
    # Convert input strings to datetime objects
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)

    # Calculate the difference using timedelta
    time_diff = end - start

    # Extract days, hours, and minutes from timedelta
    days_diff = time_diff.days
    hours_diff = time_diff.seconds // 3600
    minutes_diff = (time_diff.seconds % 3600) // 60

    return days_diff, hours_diff, minutes_diff
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
def convert_and_display_sorted_data(file_path, column_name, category_order):
    # Read the CSV file into a DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please enter a valid file path.")
        return

    # Convert the specified column to categorical with specified category order
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

    # Display sorted data based on the specified column
    sorted_data = df.sort_values(by=column_name)
    print(sorted_data)
```

Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

```python
def visualize_sales_data(file_path):
    # Read the CSV file into a DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please enter a valid file path.")
        return
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])

    # Create a pivot table for stacking product categories over time
    pivot_df = df.pivot_table(index='Date', columns='Product_Category', values='Sales', aggfunc='sum')

    # Create a stacked bar chart
    pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Sales of Product Categories Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend(title='Product Category')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.

```python
def calculate_and_display_stats(file_path):
    # Read the CSV file into a DataFrame
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please enter a valid file path.")
        return

    # Calculate mean, median, and mode of test scores
    mean_score = df['Test_Score'].mean()
    median_score = df['Test_Score'].median()
    mode_score = df['Test_Score'].mode()
    
    stats_df = pd.DataFrame({
        'Statistic': ['Mean', 'Median', 'Mode'],
        'Test_Score': [mean_score, median_score, mode_score.values[0]]  # Use .values[0] to get the mode value
    })

    # Display the results in a table
    print("\nSummary of Test Scores:")
    print(stats_df)
```
